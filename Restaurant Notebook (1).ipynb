```python
%sql
create table fullrestaurant
using delta
as
select
order_id,
cast(order_date as timestamp) as order_date,
hour,
category,
item_name,
price,
quantity,
discount,
total_amount,
branch,
payment_method,
order_type,
customer_id,
rating,
is_weekend
from restaurant_1
union all
select
order_id,
cast(order_date as timestamp) as order_date,
hour,
category,
item_name,
price,
quantity,
discount,
total_amount,
branch,
payment_method,
order_type,
customer_id,
rating,
is_weekend
from restaurant_2
union all 
select 
 order_id,
 order_date,
 hour,
 category,
 item_name,
 price,
 quantity,
 discount,
 total_amount,
 branch,
 payment_method,
 order_type,
 customer_id,
 rating,
 is_weekend
from restaurant_json_1
```
